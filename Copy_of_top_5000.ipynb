```python
import requests
import io
from bs4 import BeautifulSoup
import time
from lxml.html import fromstring, tostring
import pickle
import pandas as pd
import json
```

```python
my_headers = {
'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 ' + ' (KHTML, like Gecko) Chrome/61.0.3163.100Safari/537.36'
}

test_url = 'https://api.inc.com/rest/companyprofile/varo-bank/withlist/239'

r = requests.get(url = test_url, headers = my_headers)
print("request code: ", r.status_code)
```

```python
data = json.loads(r.text)
```

```python
companies_data = pd.DataFrame(data['fullList']['listCompanies'])
```

```python
companies_data.tail()
```

```python
urls = list(companies_data['ifc_filelocation'])
```

```python
#   function to get the each companies industry field

def get_industry(url):
    try:
        base_url = 'https://www.inc.com/profile/'
        wd.get(base_url+url)
        industry = wd.find_element_by_xpath('//*[@id="mainApp"]/div[2]/div[1]/div[2]/div[3]/div[2]/div[2]/div/div/div/div').text
        return industry
    except:
        return ''
```

## Collect the Industry data 

```python
#import time

#main()
#print("--- %s seconds ---" % (time.time() - start_time))
industries = []
i = 0
for url in urls:
     industries.append(get_industry(url))
     i +=1
     print('iteration :', i)
```

```python
industries
```

```python
#get the required columns from the companies_data DataFrame

info_df = companies_data[['rank', 'company', 'ifc_linkedin_address']]
```

```python
info_df['industry'] = industries
```
